## Separate X and Y

```python
x = df.iloc[:,:-1]
```

```python
x
```

```python
x.shape
```

```python
y = df.iloc[:,8]
```

```python
y
```

#Algorithm before split


# Train_test_split

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest= train_test_split(x,y, test_size=.2,random_state=1)
```

```python
xtrain
```

```python
ytrain
```

```python
xtest
```

# Algorithm
## RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
Rclf = RandomForestClassifier()
```

```python
Rclf.fit(xtrain,ytrain)
```

```python
scoreR= Rclf.score(xtest,ytest)
```

```python
scoreR
```

#prediction Result

```python
prediction = Rclf.predict([["72","110","55","42","60","41.2","0.441","50"]])
prediction
```

```python
if(prediction !=0):
  print('The person is diabetic')
else:
  print("The person is not diabetic")
```

#perfomance

```python
pre = Rclf.predict(xtest)
```

```python
pre
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(pre,ytest)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix (pre,ytest)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(pre,ytest))
```

```python
scoreD
```

```python
prediction = Dclf.predict([["72","110","55","42","60","41.2","0.441","50"]])
prediction
```
